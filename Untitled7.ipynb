## 1. Load Data

First, we'll load our credit card fraud dataset into a pandas DataFrame and take a quick look at the first few rows.

```python
# Import the pandas library for data manipulation
import pandas as pd

# Define the path to our CSV file
file_path = 'credit_card_fraud_10k.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first 5 rows of the DataFrame to see our data
display(df.head())
```

## 2. Data Cleaning

Now, let's clean our data by checking for missing values, duplicates, and ensuring data types are correct.

```python
# Get a summary of the DataFrame, including data types and non-null values
print('DataFrame Info:')
df.info()
```

```python
# Check for any missing values in each column
print('\nMissing Values:')
print(df.isnull())
print('\nMissing Values:')
print(df.isnull().sum())
```

```python
# Remove duplicate rows from the DataFrame
df.drop_duplicates(inplace=True)
print(f'\nNumber of rows after removing duplicates: {len(df)}')
```

```python
# Convert 'transaction_hour' to integer type for consistency
df['transaction_hour'] = df['transaction_hour'].astype(int)
```

```python
# Display info again to confirm changes
print('\nDataFrame Info after cleaning:')
df.info()
```

## 3. Basic Analysis

Let's get a general idea of our dataset's size and the distribution of fraud cases.

```python
# Show the number of rows and columns in the DataFrame
print(f'DataFrame shape (rows, columns): {df.shape}')

# Count the occurrences of fraud (1) and non-fraud (0)
print('\nFraud vs Non-Fraud Count:')
print(df['is_fraud'].value_counts())

# Calculate the percentage of fraud cases
fraud_percentage = (df['is_fraud'].sum() / len(df)) * 100
print(f'\nPercentage of Fraud Cases: {fraud_percentage:.2f}%')
```

## 4. Create Simple Graphs

We'll use `matplotlib` and `seaborn` to visualize some key aspects of our data.

```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set a style for the plots for better aesthetics
sns.set_style('whitegrid')
```

```python
# Plot 1: Fraud vs Non-Fraud Count
plt.figure(figsize=(7, 5))
sns.countplot(x='is_fraud', data=df, palette='viridis')
plt.title('Fraud vs Non-Fraud Transactions')
plt.xlabel('Is Fraud (0: No, 1: Yes)')
plt.ylabel('Number of Transactions')
plt.xticks([0, 1], ['Not Fraud', 'Fraud'])
plt.show()
```

```python
# Plot 2: Amount vs Fraud (Boxplot)
plt.figure(figsize=(8, 6))
sns.boxplot(x='is_fraud', y='amount', data=df, palette='muted')
plt.title('Transaction Amount by Fraud Status')
plt.xlabel('Is Fraud')
plt.ylabel('Transaction Amount')
plt.xticks([0, 1], ['Not Fraud', 'Fraud'])
plt.ylim(0, df['amount'].quantile(0.99)) # Limit y-axis to focus on main data
plt.show()
```

```python
# Plot 3: Transaction Hour vs Fraud (Histplot)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='transaction_hour', hue='is_fraud', multiple='stack', bins=24, palette='coolwarm')
plt.title('Transaction Hour Distribution by Fraud Status')
plt.xlabel('Transaction Hour (0-23)')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
# Plot 4: Foreign Transaction vs Fraud (Countplot)
plt.figure(figsize=(8, 5))
sns.countplot(x='foreign_transaction', hue='is_fraud', data=df, palette='cividis')
plt.title('Foreign Transactions by Fraud Status')
plt.xlabel('Foreign Transaction')
plt.ylabel('Number of Transactions')
plt.xticks([0, 1], ['Domestic', 'Foreign'])
plt.legend(title='Is Fraud', labels=['No', 'Yes'])
plt.show()
```

```python
# Plot 5: Device Trust Score vs Fraud (Boxplot)
plt.figure(figsize=(8, 6))
sns.boxplot(x='is_fraud', y='device_trust_score', data=df, palette='plasma')
plt.title('Device Trust Score by Fraud Status')
plt.xlabel('Is Fraud')
plt.ylabel('Device Trust Score')
plt.xticks([0, 1], ['Not Fraud', 'Fraud'])
plt.show()
```

### Transaction Amount Statistics

Let's look at the descriptive statistics for the 'amount' column to understand its range and distribution better. This will clarify what values are represented on the y-axis of the 'Amount vs Fraud' boxplot.

```python
# Display descriptive statistics for the 'amount' column
# This includes count, mean, standard deviation, min, max, and quartiles.
print("Descriptive statistics for 'amount' column:")
display(df['amount'].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# LinkedIn Optimization: Modern style and larger fonts
sns.set_theme(style='whitegrid', context='talk')
plt.rcParams['font.family'] = 'sans-serif'

# Create a 2x2 grid (better for 16:9 LinkedIn aspect ratio)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 12))
axes = axes.flatten()

# 1. Fraud Proportion
fraud_counts = df['is_fraud'].value_counts(normalize=True) * 100
sns.barplot(x=fraud_counts.index, y=fraud_counts.values, hue=fraud_counts.index, palette='magma', ax=axes[0], legend=False)
axes[0].set_title('Fraud Probability Distribution (%)', fontweight='bold', pad=20)
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(['Secure', 'Fraudulent'])
axes[0].set_ylabel('Percentage of Total')

# 2. Amount Distribution
sns.boxplot(x='is_fraud', y='amount', data=df, hue='is_fraud', palette='magma', ax=axes[1], showfliers=False, legend=False)
axes[1].set_title('Transaction Value: Secure vs Fraud', fontweight='bold', pad=20)
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['Secure', 'Fraud'])
axes[1].set_ylabel('USD Amount')

# 3. Peak Fraud Hours
sns.histplot(data=df, x='transaction_hour', hue='is_fraud', multiple='stack', bins=24, palette='magma', ax=axes[2], legend=False)
axes[2].set_title('Peak Hours for Fraudulent Activity', fontweight='bold', pad=20)
axes[2].set_xlabel('Hour of Day (24h)')

# 4. Impact of Foreign Transactions
sns.pointplot(x='foreign_transaction', y='is_fraud', data=df, color='#de4968', ax=axes[3])
axes[3].set_title('Foreign Transaction Risk Multiplier', fontweight='bold', pad=20)
axes[3].set_xticks([0, 1])
axes[3].set_xticklabels(['Domestic', 'International'])
axes[3].set_ylabel('Fraud Rate')

# Global Styling
plt.suptitle('DATA INSIGHT: Credit Card Fraud Trends 2024', fontsize=28, fontweight='bold', y=1.05, color='#333333')
plt.tight_layout()
plt.show()
```
